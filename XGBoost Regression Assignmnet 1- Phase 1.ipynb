```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
independent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']] 
```

```python
dependent=dataset[['charges']]
```

```python
#Train and Test dataset split 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
#Model creation
from xgboost import XGBRegressor
regressor= XGBRegressor()
regressor.fit(x_train,y_train)
```

```python
#Evaluation metrics
y_pred=regressor.predict(x_test)
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```

```python
#saving the model created
import pickle
filename="finalized_model_XGBoost.sav"
pickle.dump(regressor,open(filename,'wb'))
```

```python
#To load the saved model
loaded_model=pickle.load(open("finalized_model_XGBoost.sav",'rb'))
```

```python
#To check the model with real time inputs
result=loaded_model.predict([[35,24.6,2,True,False]])
result
```
