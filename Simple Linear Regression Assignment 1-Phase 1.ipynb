```python
#Pandas is a inbuilt library contains functions related to database
import pandas as pd
```

```python
#The csv file is imported as dataset (Data Collection)
dataset=pd.read_csv("Salary_Data.csv")
```

```python
#The columns are assigned to different variables(Input/Output Split)
independent=dataset[["YearsExperience"]]
dependent=dataset[["Salary"]]
```

```python
#From the dataset the train and test values are separated and assigned to different varibles(Train/Test split)
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
#Model Creation(Learning Phase)- With the given train dataset the AI model is learning how the graph flows
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
#To display the calculated weight values by the model created using training dataset
weight=regressor.coef_
weight
```

```python
#To display the calculated bias values by the model created using training dataset
bias=regressor.intercept_
bias
```

```python
#By using r2 Evaluation metrics method Checking whether the model created has learned correctly or not
#If the r_score value near to 1 then it is a good model(learned correctly)
y_pred=regressor.predict(x_test)
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```

```python
#To save the created model(only after Evaluation Metrics)
import pickle
filename="finalized_model_linear.sav"
pickle.dump(regressor,open(filename,'wb'))
```

```python
#To load the model created
loaded_model=pickle.load(open("finalized_model_linear.sav",'rb'))
```

```python
#To check the loaded model with real time input
result=loaded_model.predict([[13]])
result
```
